######  Q.1. You are working in a bank and you have been given two lists of the employees who worked in 2021. Employee’s name in list 1 are Ramesh, Suresh, Mahesh, Ali, Jacob and Saritha. List 2 contains the names of Ali, Mukesh, Mahesh, Jacob, Sai, Sarita. Please write a program which helps to identify people who are common in both lists. Please do not use any in-built function. 



```python
def common_string():    #DEFINING A FUNCTION
  employee_list1=['ramesh','suresh','Mahesh','Ali','Jacob','Saritha']    # PROVIDED LIST1 INFO
  employee_list2=['Ali','Mukesh','Mahesh','Jacob','Sai','Saritha']       # PROVIDED LISR2 INFO
  for i in employee_list1:                # i = ELEMENTS FOR LIST1 WITH FOR LOOP
    for j in employee_list2:              # j = ELEMENTS FOR LIST2 WITH FOR LOOP
     if i==j:                    # EQUALIZING VARIABLE'S ELEMENTS WITH IF STATEMENT
       print(i,end=" ")          # USING END FUNCTION FOR SEQUENCE RESULT AND WE CAN PRINT i OR j
common_string()                  # CALLING THE FUNCTION
```

######Q.4. You are working in a medical store. A patient came to your medical store and asked to buy 2 strips of paracetamol, 3 strips of azithromycin and 5 strips of Vitamin C. One strip of paracetamol costs Rs 35, one strip of azithromycin costs Rs 49 and one strip of vitamin c costs Rs. 33. Patient gave you Rs 2000. Please tell us what is the total cost of each medicine, total cost of all medicine and how much money you refunded to the patient? 

```python
def medicine_bill():       # DEFINING A FUNCTION 

  paracetamol=int(2*35)    # USING ARITHEMATIC OPERATOR
  azithromycin=int(3*49)   # USING ARITHEMATIC OPERATOR
  vitamin_c=int(5*33)      #  USING ARITHEMATIC OPERATOR

  print("Cost of 2 strips of paracetamol= (2x35): ",paracetamol)          # PRINTING THE RESULT
  print("Cost of 3 strips of azithromycin= (3x49): ",azithromycin)        # PRINTING THE RESULT
  print("Cost of 5 strips of vitamin_c= (5x33): ",vitamin_c)              # PRINTING THE RESULT

  k=paracetamol+azithromycin+vitamin_c       # ADDING THE VALUES INTO A VARIABLE
  print("Total medicines cost: ",k)          # PRINTING THE RESULT

  payment=2000       # GIVEN INFO
  refund=payment-k   # USING ARITHEMATIC OPERATOR 
  print("Money refuned to the patient: ",refund)                         # PRINTING THE RESULT

medicine_bill()      # CALLING THE FUNCTION
```

######Q.5. Accept a sentence as input and find the number of vowels in it. Assume that the sentence has no punctuation marks. For example: I am learning python contains 6 vowels. This function should be applicable for all other different sentences.

```python
def vowels():                 # DEFINING THE FUNCTION
 sentence=input("Enter any sentence of your choice: " )          # TAKING INPUT FROM USER
 count =0                     # SETTING COUNT TO ZERO
 vowel =set("aeiouAEIOU")          # USING SET FUNCTION 
 for letter in sentence:      # USING FOR LOOP
   if letter in vowel:       
     count=count+1            # INCREASING COUNT WITH ONE FOR EVERY TIME
 print("Count of the vowels is: ",count)
vowels()                      # CALLING THE FUNCTION 
```

######Q.6. You have been appointed by the election commission to create a website. Your first task is to work on a program which tells candidates if they are eligible for voting or not. If they are eligible your output should be ‘Congrats! You are eligible’, otherwise it should tell that you have to return after X number of years. Eligibility criteria for voting is 18 years. 
For example, If someone is 18 or above your output should be ‘Congrats! You are eligible’. If someone’s age is 15 years it should print output as ‘return after 3 years’.

```python
def election_commission():      # DEFINING THE FUNCTION
  candidate_age=int(input("Enter the candidate age: "))       # TAKING USER INPUT AND CONVERT TO INTEGER

  if candidate_age >= 18:                                     # USING IF STATEMENT
    print("Congrats! You are eligible")

  else:                                                       # USING ELSE STATEMENT
   k=18-candidate_age           # ARITHEMATIC OPERATOR
   print("Sorry! Return after", k, "years")

election_commission()           # CALLING THE FUNCTION
```
